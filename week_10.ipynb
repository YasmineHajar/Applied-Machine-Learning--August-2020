```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pup = plt.imread('images/puppy.jpg', format = 'grayscale')
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
pup2 = rgb2gray(pup)
```

```python
pup2.shape
```

```python
puppers = pup2[0:600, 200:800]
plt.imshow(puppers)
```

```python
u, s, vt = np.linalg.svd(puppers)
```

```python
s.shape
```

```python
sing_mat = np.diag(s)
```

```python
u.shape
```

```python
sing_mat.shape
```

```python
vt.shape
```

```python
sing_mat[:, 50:] = 0

approx_pup = u @ sing_mat @ vt

plt.imshow(approx_pup)
```

```python
def svd_pup(n):
    sing_mat[:, n:] = 0

    approx_pup = u @ sing_mat @ vt

    return approx_pup
```

```python
sing_mat.shape
```

```python
fig, axs = plt.subplots(2, 5, figsize = (15, 8))
svs = 10
for i in range(2):
    for j in range(5):
        axs[i, j].imshow(svd_pup(svs))
        axs[i, j].set_title(f'Singular Vals: {svs}')
        svs -= 1
        axs[i, j].axis('off')
```
