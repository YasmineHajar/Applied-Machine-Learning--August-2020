```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.countplot(penguins['species'], hue = penguins['island']);
plt.title('Species Counts by Island');
```

```python
sns.boxplot(penguins['species'], penguins['bill_length_mm'], hue = penguins['sex'], whis = 1.0);
```

```python
sns.swarmplot(penguins['species'], penguins['bill_length_mm'], alpha = 0.4, edgecolor = 'black')
sns.boxplot(penguins['species'], penguins['bill_length_mm']);
```

```python
sns.swarmplot(penguins['species'], penguins['bill_length_mm'], alpha = 0.7, edgecolor = 'black')
sns.violinplot(penguins['species'], penguins['bill_length_mm'], saturation=0.5);
```

```python
sns.swarmplot(x='species', y='bill_length_mm', data = penguins, alpha = 0.7, edgecolor = 'black')
```

```python
sns.kdeplot(penguins['bill_length_mm'], shade = True, color = 'salmon');
```

```python
plt.hist(penguins['bill_length_mm'], density = True, bins = 30, alpha = 0.2, edgecolor = 'black', color = 'firebrick');
```

```python
plt.scatter(penguins['bill_length_mm'], penguins['bill_depth_mm'])
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = penguins)
```

```python
sns.pairplot(penguins, hue = 'species')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 6))
sns.boxplot('species', 'bill_depth_mm', data = penguins, hue = 'sex', ax = ax[0])
sns.scatterplot('bill_depth_mm', 'flipper_length_mm', data = penguins,  ax = ax[1])
```

```python
sns.regplot('bill_depth_mm', 'flipper_length_mm', data = penguins)
plt.savefig('imname.png')
```
