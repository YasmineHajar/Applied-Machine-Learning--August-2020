```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
bdog_grey = rgb2gray(bdog)
```

```python
plt.imshow(bdog_grey, cmap = 'gray')
```

```python
plt.imshow(bdog_grey.T, cmap = 'gray')
```
